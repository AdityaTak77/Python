#### list comprehension
basic syntax [exrpession for item in iterable]
with conditional logic [ if condition] nested list [for item in iterable1 for item2 in iterable2]

```python
### basic

square= [num**2 for num in range(10)]
square
```

```python
### with condition
lst=[]
for i in range(10):
  if i%2==0:
    lst.append(i)
print(lst)
```

```python
even_numbers= [num for num in range(10) if num%2==0]
print(even_numbers)
```

```python
### Nested list comprehension
lst1=[1,2,3,4]
lst2=['a','b','c','d']
pair= [[i,j] for i in lst1 for j in lst2]
pair
```

```python
### list comprehension with fucntion calls
words = ["hello","world","python"]
length = [len(word) for word in words]
length
```

```python
### Nested dict

s= {
    "student1":{"name":"muna","age":32},
    "student2":{"name":"king","age":77},
    "student3":{"name":"sing","age":35},
}
s
```

```python
## access nested dict elements

print(s["student1"]["name"])
```

```python
s.items()
```

```python
## iterating over nested dict

for student_id,student_info in s.items():
  print(f"{student_id}:{student_info}")
  for key, value in student_info.items():
    print(f"{key}:{value}")
```

```python
## Practical Example

## use dict to count the frequency of elements in list

n = [1,1,2,3,4,4,55,6,7,78,8,8,9,9,9,9]

freq = {}

for num in n:
  if num in freq:
    freq[num]+=1
  else:
    freq[num]=1
print(freq)
```

## Real world examples

```python
# Manage a do to list

todo=["buy items","clean","study"]
## add task
todo.append("run")
todo.append("walk")
## remove a completed task
todo.remove("clean")
## checking the task
if "study" in todo:
  print("kindly do it")
print("remaining")
for task in todo:
  print(f"-{task}")
```

```python
# Organizing student grades
grades = [85, 92, 78, 90, 88]

# Adding a new grade
grades.append(95)

# Calculating the average grade
average_grade = sum(grades) / len(grades)
print(f"Average Grade: {average_grade:.2f}")

# Finding the highest and lowest grades
highest_grade = max(grades)
lowest_grade = min(grades)
print(f"Highest Grade: {highest_grade}")
print(f"Lowest Grade: {lowest_grade}")
```

```python
# Managing an inventory
inventory = ["apples", "bananas", "oranges", "grapes"]

# Adding a new item
inventory.append("strawberries")

# Removing an item that is out of stock
inventory.remove("bananas")

# Checking if an item is in stock
item = "oranges"
if item in inventory:
    print(f"{item} are in stock.")
else:
    print(f"{item} are out of stock.")

# Printing the inventory
print("Inventory List:")
for item in inventory:
    print(f"- {item}")
```

```python
# Collecting user feedback
feedback = ["Great service!", "Very satisfied", "Could be better", "Excellent experience"]

# Adding new feedback
feedback.append("Not happy with the service")

# Counting specific feedback
positive_feedback_count = sum(1 for comment in feedback if "great" in comment.lower() or "excellent" in comment.lower())
print(f"Positive Feedback Count: {positive_feedback_count}")

# Printing all feedback
print("User Feedback:")
for comment in feedback:
    print(f"- {comment}")
```
