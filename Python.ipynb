#### list comprehension
basic syntax [exrpession for item in iterable]
with conditional logic [ if condition] nested list [for item in iterable1 for item2 in iterable2]

```python
### basic

square= [num**2 for num in range(10)]
square
```

```python
### with condition
lst=[]
for i in range(10):
  if i%2==0:
    lst.append(i)
print(lst)
```

```python
even_numbers= [num for num in range(10) if num%2==0]
print(even_numbers)
```

```python
### Nested list comprehension
lst1=[1,2,3,4]
lst2=['a','b','c','d']
pair= [[i,j] for i in lst1 for j in lst2]
pair
```

```python
### list comprehension with fucntion calls
words = ["hello","world","python"]
length = [len(word) for word in words]
length
```
