```python
###fucntion with multiple parameters

def add(a,b):
  return a+b
result=add(2,4)
print(result)
```

```python
## default parameters

def greet(name):
  print(f"hell0 {name} welcome " )
greet("muna")
```

```python
def greet(name="guest"):
  print(f"hell0 {name} welcome " )
greet("")
```

```python
### variable length arguments
## positional and keywords arguments

def printnumber(*args):
  for n in args:
    print(n)
```

```python
printnumber(1,2,3,4,5,6,7,"args")
```

```python
### keywords arguments all pairs will be in key value pairs

def printdetails(**kwargs):
  for key, value in kwargs.items():
    print(f"{key}:{value}")

printdetails(nmae="muna",age="32")
```

```python
def printdetails(*args,**kwargs):
  for val in args:
    print(f"positional argument:{val}")


  for key, value in kwargs.items():
    print(f"{key}:{value}")

printdetails(1,2,3,4,"muna", name="muna",age="32")
```
