## Functions examples

```python
## Ex1 Temperatures conversion
def convert (temp, unit):
  """this function converts temperature from celsius to fahrenheit"""
  if unit== 'c':
    return temp *9/5 +32 ## celcius to fahrenheit
  elif unit=='f':
      return (temp-32)* 5/9 ## fahrenheit to celcius
  else :
      return none
convert (25,'c')
```

```python
## password strength checker

def check (password):
  """this function check if the password is strong or not"""
  if len(password)<8:
    return False
  if not any(char.isdigit() for char in password):
    return False
  if not any(char.isupper() for char in password):
    return False
  if not any(char.islower() for char in password):
    return False
  if not any(char in '!@#$%^&*' for char in password):
    return False
  return True
check("muna@123")
```

```python
## calculate the total cost of items in a shopping cart

def cal (cart):
  totalcost=0
  for item in cart:
    totalcost+=item['price']*item['quantity']
  return totalcost


## example cart data
cart=[
    {'name':'apple','price':2.5,'quantity':3},
    {'name':'banana','price':1.5,'quantity':2},
]

## calling the function

totalcost = cal(cart)
print(totalcost)
```

```python
## check if a string is palindrome
## example: ABA

def pali (s):
  s=s.lower().replace(" ","")
  return s==s[::-1]
print(pali("ABA"))
print(pali("a man nama"))
```

```python
## calculate the factorial of a number using recursion

def factor(n):
  if n==0:
    return 1

  else :
    return n * factor(n-1)
print(factor(6))
```
